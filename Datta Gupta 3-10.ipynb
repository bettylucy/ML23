```python
##test matrix population
import random
import numpy as np

numbers = []
for i in range (0,100):
    jrow=[]
    for j in range (0,10):
        jrow.append(random.random())
    numbers.append(jrow)    

for m in range(0,100):  
    for k in range(0,10):  
        print(numbers[m][k], end=" ")  
    print()
```

```python
#statistics test
import statistics
import numpy as np

numbersarray=np.array(numbers)
numbersmean=[]

for i in range (0,10):
    imean=statistics.mean(numbersarray[:,i])
    numbersmean.append(imean)
    
print(numbersmean)
#print(numbersarray[:,i])

```

```python
import numpy as np
from numpy import random

numbersraw = random.rand(100,10)

print(numbersraw)
```

```python
print(numbersraw[:,0])
```

```python
##calculating moments
import statistics

meanlist=[]
variancelist=[]

for i in range (0,10):
    imean=statistics.mean(numbersraw[:,i])
    meanlist.append(imean)
    ivar=statistics.variance(numbersraw[:,i])
    variancelist.append(ivar)
    
print(meanlist)
print(variancelist)
```

```python
##plotting

from matplotlib import pyplot as plt

plt.hist(meanlist, 10)

plt.show()
```
